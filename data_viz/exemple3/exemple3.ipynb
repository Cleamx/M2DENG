```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi, 50)
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, np.sin(x))
ax[0, 1].plot(x, np.cos(x))
ax[1, 0].plot(x, np.tan(x))
ax[1, 1].plot(x, np.exp(x))

plt.tight_layout()
plt.show()
```

```python
categories = ['category 1', 'category 2', 'category 3', 'category 4']
values = [80, 20, 50, 30]
fig, ax1 = plt.subplots()
ax1.bar(categories, values, color='skyblue')
ax1.set_ylabel('Values', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
cumulative_values = [sum(values[:i+1]) for i in range(len(values))]
ax2.plot(categories, cumulative_values, color='red', marker='o')
ax2.set_ylabel('Cumulative Values', color='red')
ax2.tick_params(axis='y', labelcolor='red')
plt.title('Pareto chart')
plt.show()
```

```python
labels = ('Python', 'Java', 'Scala', 'C#')
sizes = [45, 20, 15, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'Distribution 1': np.random.randn(100),
    'Distribution 2': np.random.randn(100) * 2,
    'Distribution 3': np.random.randn(100) * 3
})

sns.boxplot(data=df)
plt.title('Box plot for three distributions')
plt.show()
```
